## Question 3: How many students who are enrolled in a computing degree/certificate were influenced positively by someone not employed in CCM (HS Guidance Counselor, Teacher, Alumni, Employer, current CCM student) and how has that total number and percentage of students enrolled in said programs changed from Fall 2020 - Fall 2024?

#### Need to make sure the data is being pulled from both cleaned data files since this will be a comparison of the rate of change between the two

```python
df_Fall_2020 = pd.read_csv("Fall_2020 Cleaned.csv")
df_Fall_2024 = pd.read_csv("Fall_2024 Cleaned.csv")
```

#### Using a dictionary so that the x-axis labels can be amended since the columns at this point were not as intuitive because of similar types of questions later in the survey.

```python
positive_influences = {
    'Comp_Friend_Family': 'Friend/Family',
    'Comp_Personal_Ref': 'Personal_Referral',
    'Comp_Teacher_Counselor': 'Teacher/Counselor',
    'Comp_Employer': 'Employer',
    'Comp_Current_Student': 'Current_CCM_Student'
}
```

#### The code was different here for the function being used since it was pulling from a dictionary and not a list. In addition, there was a need to convert the value to a percentage, and the 'count' was inclusive of any impact (some and high were not differentiated).

```python
def calc_impact_percentages(df, columns_dict):
    result = {}
    total = len(df)
    for col, label in columns_dict.items():
        count = df[col].isin(['Some Impact', 'High Impact']).sum()
        result[label] = count / total * 100
    return result
```

```python
impact_2020 = calc_impact_percentages(df_Fall_2020, positive_influences)
impact_2024 = calc_impact_percentages(df_Fall_2024, positive_influences)
```

#### Identifying labels an setting the width of the bars prior to plotting

```python
labels = list(positive_influences.values())
x = np.arange(len(labels))
width = 0.30
```

```python
values_2020 = [impact_2020[label] for label in labels]
values_2024 = [impact_2024[label] for label in labels]
```

```python
plt.figure(figsize=(12, 7))
plt.bar(x - width/2, values_2020, width, label='Fall 2020', color='gold')
plt.bar(x + width/2, values_2024, width, label='Fall 2024', color='dodgerblue')
plt.xticks(x, labels, rotation=45, ha='right')
plt.ylabel('Percentage of Students (%)')
plt.title('External Influences on Computing Enrollment by Source', fontsize = 18, color = 'darkblue')
plt.ylim(0, 100)
plt.legend(loc = 'upper left', fontsize = 14)
```

### Findings: In general, those who know the students on a more personal level seem to have the greatest influence over students' decisions to enroll in computing courses. In total, the number of these referring sources seems to show an increase from Fall 2020 to Fall 2024, with the exception of Teacher/Counselor influences and Current CCM Students.
